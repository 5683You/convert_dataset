```python
import json
import time
from datasets import load_dataset

# 加载数据集
dataset = load_dataset('AdaptLLM/finance-tasks', 'Headline')['test']

# 记录开始时间
start_time = time.time()

# 初始化结果列表
results = []

# 解析每个条目
for index, entry in enumerate(dataset):
    input_string = entry['input']
    options = entry['options']
    gold_index = entry['gold_index']
    
    # 获取正确答案
    answer = options[gold_index]
    
    # 将输入字符串拆分为问题和答案
    questions = input_string.split('\n')
    
    for i, question in enumerate(questions):
        if question.strip():  # 确保问题不为空
            result = {
                "id": f"{index}_{i}",
                "Question": question.strip(),
                "Answer": answer.strip()
            }
            results.append(result)

# 保存为 JSON 文件
output_file = 'reformatted_headline_dataset.json'
with open(output_file, 'w') as f:
    json.dump(results, f, indent=4)

# 记录结束时间
end_time = time.time()

# 打印统计信息
print(f"Total Q&A pairs: {len(results)}")
print(f"Time taken: {end_time - start_time} seconds")
print(f"Results saved to {output_file}")
```
